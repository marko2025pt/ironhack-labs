## Exercise 1

```python
# Step 1: Define the greeting function
# -------------------------------------
# We are creating a function called greet_person that takes two parameters:
# - name: the name of the person (required)
# - language: the favorite programming language (optional, default is "Python")
# Parameters:
#    name (str): Name of the person
#    language (str, optional): Favorite programming language, defaults to "Python"
# Returns:
#    str: Personalized greeting message
 
def greet_person(name, language="Python"):

    # Step 2: Create the base greeting message using the name
    message = f"Hello, {name}! Welcome to Python programming!"
    
    # Step 3: Check if the language is provided and is not the default "Python"
    if language != "Python":
        # Step 4: If a different language is provided, append this info to the message
        message += f" I see you're interested in {language} too!"
    
    # Step 5: Return the final greeting message
    return message

# Step 6: Test the function with different inputs
# -----------------------------------------------

# Test case 1: Only name provided, should use default language "Python"
print(greet_person("Alice"))

# Test case 2: Name and a different language provided, should mention JavaScript
print(greet_person("Bob", "JavaScript"))

# Test case 3: Name and default language provided, should not mention the language
print(greet_person("Charlie", "Python"))
```

## Exercise 2


```python
# Step 1: Define the grade calculator function
# --------------------------------------------
# We are creating a function called calculate_grade that takes one parameter:
# - percentage: the numeric score of the student (required)
# Parameters:
#    percentage (float or int): Score between 0 and 100
# Returns:
#    str: Letter grade ("A", "B", "C", "D", "F")
def calculate_grade(percentage):
    # Step 2: Conditional logic to assign grades
    # 90 or above => A
    if percentage >= 90:
        return "A"
    # 80–89 => B
    elif percentage >= 80:
        return "B"
    # 70–79 => C
    elif percentage >= 70:
        return "C"
    # 60–69 => D
    elif percentage >= 60:
        return "D"
    # Below 60 => F
    else:
        return "F"

# ------------------------------
# Step 3: Test the function with provided values
# ------------------------------

test_scores = [95, 85, 75, 65, 55, 90, 60]

# Loop through test scores and print the results
for score in test_scores:
    print(f"Score: {score} → Grade: {calculate_grade(score)}")
```

```python
# Step 1: Define the grade calculator function with input validation
# ------------------------------------------------------------------
# We are creating a function called calculate_grade that takes one parameter:
# - percentage: the numeric score of the student (required)
# The function now also checks if the percentage is valid (between 0 and 100)
# Parameters:
#    percentage (float or int): Score between 0 and 100
# Returns:
#    str: Letter grade ("A", "B", "C", "D", "F") or error message if invalid
def calculate_grade(percentage):
    
    # Step 2: Validate input
    # Check if the percentage is outside the valid range
    if percentage < 0 or percentage > 100:
        return "Invalid percentage. Please enter a value between 0 and 100."
    
    # Step 3: Conditional logic to assign grades
    # 90 or above => A
    if percentage >= 90:
        return "A"
    # 80–89 => B
    elif percentage >= 80:
        return "B"
    # 70–79 => C
    elif percentage >= 70:
        return "C"
    # 60–69 => D
    elif percentage >= 60:
        return "D"
    # Below 60 => F
    else:
        return "F"

# ------------------------------
# Step 3: Test the function with provided values, including not valid vaules
# ------------------------------

test_scores = [95, 85, 75, 65, 55, 90, 60,-5, 105]

# Loop through test scores and print the results
for score in test_scores:
    print(f"Score: {score} → Grade: {calculate_grade(score)}")
```

## 1. What is the main advantage of using functions instead of writing the same code multiple times?
The main advantage of using functions is **code reusability**. A function lets us write logic once and reuse it many times with different inputs. This makes the code shorter and easier to maintain. If something needs to change, I only need update the function instead of fixing the same code in several places.

## 2. In Exercise 1, why did we use a default parameter value for `language`?
We used a default parameter value so the function could work even when the user does not provide a language. This makes the function more flexible and easier to use, because the caller only needs to pass the required information (the name), and it avoids errors. 

## 3. In Exercise 2, why is it important to check the conditions in order from highest to lowest (90, then 80, then 70, etc.)?
In this case conditions are evaluated from top to bottom. When checking from highest to lowest, each grade range is clearly separated and higher scores are matched first, which avoids incorrect classifications.

It is also possible to check from lowest to highest, but in that case the conditions must be written differently (for example, using upper bounds like `percentage < 60`, `percentage < 70`, etc.). **See example below** The **key point** is that the order and the comparison logic must be consistent.
